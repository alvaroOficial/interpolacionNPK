```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go

# Carga de datos
data = pd.read_excel("/content/EJEMPLO.xlsx")

# Definir las coordenadas X e Y para la cuadrícula
cordx = np.arange(min(data["Longitud"]), max(data["Longitud"]), 0.0001)  # Ajusta el paso para mejorar la resolución
cordy = np.arange(min(data["Latitud"]), max(data["Latitud"]), 0.0001)

# Crear la matriz para la interpolación
matz = np.zeros((len(cordy), len(cordx)))

# Método de distancia inversa
for i in range(len(cordx)):
    for j in range(len(cordy)):
        distancia = 1 / (np.sqrt((data["Latitud"] - cordy[j])**2 + (data["Longitud"] - cordx[i])**2))
        indice = np.isinf(distancia)
        if not indice.any():
            factor = distancia / np.sum(distancia)
            matz[j, i] = np.sum(data["N"] * factor)
        else:
            matz[j, i] = data["N"][indice.argmax()]

# Gráfico de contorno
fig = go.Figure()

# Capa de contorno
fig.add_trace(go.Contour(
    z=matz,
    x=cordx,
    y=cordy,
    contours=dict(start=0, end=max(data["N"]), size=1, showlabels=False),
    line_smoothing=1
))

# Capa de los puntos originales
fig.add_trace(go.Scatter(
    x=data["Longitud"],
    y=data["Latitud"],
    mode="markers",
    marker=dict(color="black", size=10),
    name="Puntos originales"
))

# Configuración del diseño
fig.update_layout(
    width=1000,
    height=800,
    title="Interpolación de Nitrógeno",
    xaxis_title="Longitud",
    yaxis_title="Latitud"
)

# Mostrar el gráfico
fig.show()
```

```python
# Función para interpolar el valor de N en unas coordenadas dadas
def obtener_valor_interpolado(latitud, longitud, data):
    distancia = 1 / (np.sqrt((data["Latitud"] - latitud)**2 + (data["Longitud"] - longitud)**2))
    indice = np.isinf(distancia)
    if not indice.any():
        factor = distancia / np.sum(distancia)
        valor_interpolado = np.sum(data["N"] * factor)
    else:
        valor_interpolado = data["N"][indice.argmax()]

    return valor_interpolado

# Ejemplo de uso: ingresar unas coordenadas y obtener el valor interpolado de N
latitud_consulta = 1.215  # Cambia a la latitud de consulta
longitud_consulta = -77.284 # Cambia a la longitud de consulta

valor_interpolado = obtener_valor_interpolado(latitud_consulta, longitud_consulta, data)
print(f"El valor interpolado de N en ({latitud_consulta}, {longitud_consulta}) es: {valor_interpolado}")
```
